```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, coint
from statsmodels.tsa.arima.model import ARIMA
from google.colab import files
import io
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

# 1. Cargar los datos desde el archivo Excel
def load_data(file_content):
    excel_file = pd.ExcelFile(io.BytesIO(file_content))
    sheets = excel_file.sheet_names
    print("\nHojas disponibles en tu Excel:", sheets)

    companies = ['META', 'AMZN', 'NVIDIA', 'VISA', 'WMT']
    timeframes = ['5MIN', '30MIN']
    target_sheets = {
        'META_5MIN': 'META 5 min',
        'META_30MIN': 'META 30 MIN',
        'AMZN_5MIN': 'AMZN 5 min',
        'AMZN_30MIN': 'AMZN 30 min',
        'NVIDIA_5MIN': 'NVIDIA 5 min',
        'NVIDIA_30MIN': 'NVIDIA 30 min',
        'VISA_5MIN': 'VISA 5 MIN',
        'VISA_30MIN': 'VISA 30 MIN',
        'WMT_5MIN': 'WMT 5 MIN',
        'WMT_30MIN': 'WMT 30 MIN'
    }

    data = {}
    for key, sheet in target_sheets.items():
        if sheet in sheets:
            # Leer sin asumir encabezado y asignar nombres manualmente
            df = pd.read_excel(excel_file, sheet_name=sheet, header=None)
            # Asignar nombres esperados
            if len(df.columns) >= 3:  # Asegurarse de que hay al menos 3 columnas
                df.columns = ['Local Date', 'Local Time', 'Close']  # Asignar nombres esperados
                # Convertir 'Close' a numérico
                df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
                # Invertir el orden para ir de más antiguo a más reciente
                data[key] = df['Close'][::-1].reset_index(drop=True)
                print(f"Hoja '{sheet}' cargada como {key}. Filas: {len(data[key])}")
                # Imprimir primeras filas para inspección
                print(f"Primeras 5 filas de '{sheet}':\n{df.head()}")
            else:
                print(f"¡Advertencia! La hoja '{sheet}' no tiene suficientes columnas. Columnas: {list(df.columns)}")
        else:
            print(f"¡Advertencia! La hoja '{sheet}' no existe en el Excel.")
    return data

# 2. Prueba de raíz unitaria (ADF)
def unit_root_test(series, name):
    result = adfuller(series.dropna())
    print(f'\nPrueba ADF para {name}:')
    print(f'Estadístico ADF: {result[0]}')
    print(f'p-valor: {result[1]}')
    print(f'Valores críticos: {result[4]}')
    return result[1] < 0.05  # True si es estacionaria

# 3. Ajustar modelo ARMA con selección automática y pronóstico
def fit_arma(series, name, max_p=3, max_q=3, forecast_steps=10):
    best_aic = float('inf')
    best_order = None
    best_model = None

    series = series.dropna()
    stationary = unit_root_test(series, name)
    if not stationary:
        series_diff = series.diff().dropna()
        print(f'{name} no es estacionaria, se aplicó diferenciación')
    else:
        series_diff = series
        print(f'{name} es estacionaria')

    # Visualización de la serie original y diferenciada
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(series, label='Serie original')
    plt.title(f'Serie original - {name}')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(series_diff, label='Serie diferenciada', color='orange')
    plt.title(f'Serie diferenciada - {name}')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Selección del mejor modelo ARMA
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                model = ARIMA(series_diff, order=(p, 0, q))
                results = model.fit()
                if results.aic < best_aic:
                    best_aic = results.aic
                    best_order = (p, 0, q)
                    best_model = results
            except:
                continue

    if best_model is not None:
        print(f'\nMejor modelo ARMA para {name}:')
        print(f'Orden: {best_order}')
        print(f'AIC: {best_aic}')

        # Visualización de residuos
        plt.figure(figsize=(12, 4))
        plt.plot(best_model.resid, label='Residuos')
        plt.title(f'Residuos del modelo ARMA - {name}')
        plt.legend()
        plt.show()

        # Pronóstico
        forecast = best_model.forecast(steps=forecast_steps)
        forecast_index = range(len(series_diff), len(series_diff) + forecast_steps)

        plt.figure(figsize=(12, 6))
        plt.plot(series_diff, label='Serie diferenciada')
        plt.plot(forecast_index, forecast, label='Pronóstico', color='red')
        plt.title(f'Pronóstico de {forecast_steps} pasos para {name}')
        plt.legend()
        plt.show()

        print(f'\nPronóstico de {forecast_steps} pasos para {name}:')
        print(forecast)
    else:
        print(f'No se pudo ajustar un modelo ARMA para {name}.')

    return best_model

# 4. Prueba de cointegración
def cointegration_test(series1, series2, name1, name2):
    series1 = pd.Series(series1, dtype='float64').dropna()
    series2 = pd.Series(series2, dtype='float64').dropna()

    # Alinear las series al tamaño de la más corta
    series1, series2 = series1.align(series2, join='inner')

    if len(series1) == 0 or len(series2) == 0:
        print(f"\nAdvertencia: No hay datos suficientes para la prueba de cointegración entre {name1} y {name2}.")
        return False

    print(f"Longitud de {name1} después de alinear: {len(series1)}")
    print(f"Longitud de {name2} después de alinear: {len(series2)}")

    try:
        score, p_value, _ = coint(series1, series2)
        print(f'\nPrueba de cointegración entre {name1} y {name2}:')
        print(f'p-valor: {p_value}')
        return p_value < 0.05  # True si hay cointegración
    except ValueError as e:
        print(f"\nError en la prueba de cointegración entre {name1} y {name2}: {e}")
        return False

# 5. Función principal
def main():
    print("Por favor, sube tu archivo Excel:")
    uploaded = files.upload()

    file_name = list(uploaded.keys())[0]
    file_content = uploaded[file_name]
    data = load_data(file_content)

    companies = ['META', 'AMZN', 'NVIDIA', 'VISA', 'WMT']
    timeframes = ['5MIN', '30MIN']

    results = {}

    # Análisis individual para cada serie
    for sheet, series in data.items():
        print(f'\n=== Análisis para {sheet} ===')
        model = fit_arma(series, sheet, forecast_steps=10)  # Pronosticar 10 pasos
        results[sheet] = model

    # Pruebas de cointegración entre pares
    for i, stock1 in enumerate(companies):
        for stock2 in companies[i+1:]:
            for timeframe in timeframes:
                series1 = data.get(f'{stock1}_{timeframe}')
                series2 = data.get(f'{stock2}_{timeframe}')
                if series1 is not None and series2 is not None:
                    cointegration_test(series1, series2,
                                     f'{stock1}_{timeframe}',
                                     f'{stock2}_{timeframe}')

if __name__ == "__main__":
    main()
```
